```python
# Import dependencies
import pandas as pd
# For data base creation
from sqlalchemy import create_engine
from config import db_password
```

```python
# Define directory path for files to keep from retyping every time
file_dir = 'Data/'
file_dir_R = 'Data/Resources/'
```

```python
# creating a player_id dataframe set to the value of atp_player csv
player_id = pd.read_csv(f'{file_dir_R}atp_players.csv', low_memory=False)
player_id.head()
```

```python
# Extracting our selected players from player_id and placing them into a dataFrame called the_big_five
the_big_five = pd.DataFrame()
the_big_five = the_big_five.append(player_id[(player_id.name_last == 'Federer') & (player_id.name_first == 'Roger')])
the_big_five = the_big_five.append(player_id[(player_id.name_last == 'Wawrinka') & (player_id.name_first == 'Stan')])
the_big_five = the_big_five.append(player_id[(player_id.name_last == 'Nadal') & (player_id.name_first == 'Rafael')])
the_big_five = the_big_five.append(player_id[(player_id.name_last == 'Murray') & (player_id.name_first == 'Andy')])
the_big_five = the_big_five.append(player_id[(player_id.name_last == 'Djokovic') & (player_id.name_first == 'Novak')])

the_big_five = the_big_five.reset_index(drop=True)
the_big_five.head()
```

```python
 # creating the Tennis Database
db_string = f'postgresql://postgres:{db_password}@127.0.0.1:5432/tennis_data'
engine = create_engine(db_string)
the_big_five.to_sql(name='players', con=engine, if_exists='replace')
```
